### DATA CLEANING

### Loading and Inspecting the datset

```python
import pandas as pd
df = pd.read_csv('DelayedFlights.csv')
df.info()
df.head()
```

### Checking the shape of the dataset

```python
df.shape
```

### Checking the number of NaNs in every column

```python
print(df.isna().sum())
```

```python
df = df.dropna(subset = ['ArrTime', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime', 'ArrDelay', 'TaxiIn', 'TaxiOut', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay'])
```

```python
df.shape
```

```python
print(df.isna().sum())
```

```python
df.head()
```

### checking for duplicate rows in the dataset

```python
print(df.duplicated().sum())
```

### Removing the columns which are not required

```python
df = df.drop(columns = ['TailNum', 'Unnamed: 0'])
```

```python
df.info()
```

### Correcting the Datatypes of DepTime and ArrTime

```python
df['DepTime'] = df['DepTime'].astype(int)
df['DepHour'] = df['DepTime'] // 100

df['ArrTime'] = df['ArrTime'].astype(int)
df['ArrHour'] = df['ArrTime'] // 100
```

```python
df.head()
```

### Cancellation code standardization

```python
print(df['CancellationCode'].unique())
```

### Checking if non of the flight od cancelled and diverted

```python
print(df['Diverted'].unique())
```

```python
print(df['Cancelled'].unique())
```

### removing extra spaces from categorical columns!

```python
df['UniqueCarrier'] = df['UniqueCarrier'].str.strip()
df['Origin'] = df['Origin'].str.strip()
df['Dest'] = df['Dest'].str.strip()
```

### Checking for outliers in numeric columns

```python
df[['ArrDelay', 'DepDelay']].describe()
```

### Plotting Boxplot to check outliers in DepDelay

```python
import matplotlib.pyplot as plt 
import seaborn as sns

plt.figure(figsize = (10,5))
sns.boxplot(df['DepDelay'])
plt.show()
```

```python
df[['DepDelay','ArrDelay']].head()
```

```python
original_df = df
```

### EDA (Exploratory Data Analysis)

### Airline wise average arrival and departure delay 

```python
airline_delay = filtered_df.groupby('UniqueCarrier')[['ArrDelay', 'DepDelay']].mean().reset_index()
print(airline_delay)
```

```python
# B6 airlines have avg arrival delay of 78 mins and avg departure delay of 72 mins!
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Airline-wise avg ArrDelay
plt.figure(figsize=(12,6))
sns.barplot(data=airline_delay, x='UniqueCarrier', y='ArrDelay',color = 'g')
plt.title("Average Arrival Delay per Airline")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Airline-wise avg DepDelay
plt.figure(figsize=(12,6))
sns.barplot(data=airline_delay, x='UniqueCarrier', y='DepDelay',color ='teal')
plt.title("Average Departure Delay per Airline")
plt.xticks(rotation=45)
plt.show()
```

### Month wise Delay pattern

```python
month_delay = filtered_df.groupby('Month')[['ArrDelay', 'DepDelay']].mean().reset_index()
print(month_delay)
```

```python
# December month has a maximum avg of arrival delay.
# july month has a maximum avg of departure delay.
# october month has a minimum avg of arrival delay.
# october month has a minimum of avg departure delay.
```

### Origin vs destination-wise delay

```python
origin_delay = filtered_df.groupby('Origin')[['ArrDelay', 'DepDelay']].mean().reset_index()
dest_delay = filtered_df.groupby('Dest')[['ArrDelay', 'DepDelay']].mean().reset_index()
```

```python
print(origin_delay)
```

```python
print(origin_delay.sort_values(by = 'ArrDelay', ascending = False))
```

```python
print(origin_delay.sort_values(by = 'DepDelay', ascending = False))
```

```python
top_airports = origin_delay.sort_values(by='ArrDelay', ascending=False).head(15)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
sns.barplot(y='Origin', x='ArrDelay', data=top_airports,color = 'blue')
plt.title('Top 15 Origins with Highest Average Arrival Delay')
plt.xlabel('Average Arrival Delay (minutes)')
plt.ylabel('Origin Airport')
plt.show()
```

```python
top_airports = origin_delay.sort_values(by='DepDelay', ascending=False).head(15)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
sns.barplot(y='Origin', x='DepDelay', data=top_airports,color = 'y')
plt.title('Top 15 Origins with Highest Average destination Delay')
plt.xlabel('Average destination Delay (minutes)')
plt.ylabel('Origin Airport')
plt.show()
```

```python
top_airports = dest_delay.sort_values(by='DepDelay', ascending=False).head(15)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
sns.barplot(y='Dest', x='DepDelay', data=top_airports,color = 'black')
plt.title('Top 15 destinations with Highest Average departure Delay')
plt.xlabel('Average departure Delay (minutes)')
plt.ylabel('Destination Airport')
plt.show()
```

```python
# CMX has the highest average arrival delay as an origin airport (137 mins).
# ACY has the highest average departure delay as an origin airport (131 mins).
# ALO has the highest average arrival delay as a destination (89 mins) and departure delay as a destination (82 mins).
```

### Hour of day impact

```python
filtered_df['DepHour'] = filtered_df['CRSDepTime'] // 100
hour_delay = filtered_df.groupby('DepHour')[['ArrDelay', 'DepDelay']].mean().reset_index()
print(hour_delay)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(hour_delay['DepHour'], hour_delay['ArrDelay'], label='Arrival Delay', marker='o')
plt.plot(hour_delay['DepHour'], hour_delay['DepDelay'], label='Departure Delay', marker='o')``
plt.xticks(range(0, 24))
plt.title('Average Delays by Departure Hour')
plt.xlabel('Hour of Departure')
plt.ylabel('Average Delay (minutes)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Max average delay hour (Departure): 4 AM
# Max average delay hour (Arrival): Also around 4–5 AM
# Least delays: Night hours (21–23), and mid-morning hours (9–11).
```

### Delay type breakdown

```python
delay_cols = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
delay_means = filtered_df[delay_cols].mean()
print(delay_means)
```

```python
# Major causes of flight delays are late-arriving aircraft and carrier-related operational issues, contributing ~44 minutes combined on average. 
# Surprisingly, weather and security delays are relatively minor contributors."
```
